## 3.

Create a variable `ingredients` with a Series that looks like:

```

Milk       1 cup

Spam       1 can
Name: Dinner, dtype: object
```

````python
# Creating a series through dictionary input
# This answer is techincally correct - just a different way of completing the task
d = {'Flour':'4 cups', 'Milk':'1 cup', 'Eggs':'2 large', 'Spam':'1 can'}
ingredients = pd.Series(data = d, index=['Flour', 'Milk', 'Eggs', 'Spam'], name = "Dinner")

# Check your answer
q3.check()
ingredients
````
